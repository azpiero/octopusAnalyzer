```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import glob
import csv
%matplotlib inline
```

```python
def pil2cv(image):
    new_image = np.array(image,dtype=np.uint8)
    if new_image.ndim == 2:
        pass
    elif new_image.shape[2] == 3:
        new_image = new_image[:,:,::-1]
    elif  new_image.shape[2] == 4:
        new_image = new_image[:,:,[2,1,0,3]]
    return new_image
```

```python
def culculate_each_area(img):
    threshold = 10.0
    gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    counters, hierarchy = cv2.findContours(gray, cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    for i ,cnt in enumerate(counters):
        # 輪郭の面積を計算する
        area = cv2.contourArea(cnt)
        if(area > threshold):
            pass
            # print("conter:{},area:{}".format(i,area))
```

```python
# Yellowヒートマップ作成テスト

folder = "/Users/takeru/Documents/django/octopusReference/octpas/result/"

img_pil = Image.open("/Users/takeru/Documents/django/octopusReference/octpas/adjuststack.tif")
img_length = img_pil.n_frames
# nparray
a = np.zeros((img_pil.size[0],img_pil.size[1]))

for i in range(0,img_length):
    img_pil.seek(i)
    img = img_pil.copy()
    #plt.imshow(img)
    img_cv2 = pil2cv(img)
    
    img_rgb = cv2.cvtColor(img_cv2, cv2.COLOR_BGR2RGB)
    

    # RGBの色抽出下限
    rgb_lower_yellow =np.array([ 100, 100, 0])
    # RGBの色抽出上限
    rgb_upper_yellow =np.array([ 200, 200, 100])
    img_mask_yellow = cv2.inRange(img_rgb, rgb_lower_yellow, rgb_upper_yellow)
    result_yellow = cv2.bitwise_and(img_rgb,img_rgb,mask=img_mask_yellow)
    # culculate_each_area(result_yellow)

    img_gray = cv2.cvtColor(result_yellow, cv2.COLOR_RGB2GRAY)
    
    # ちっこい島みたいな情報(ゴミデータ)は消す(前処理)
    # 第三引数(karnel) : 
    kernel = np.ones((5,5),np.uint8)
    opening =cv2.morphologyEx(img_gray,cv2.MORPH_OPEN,kernel)
    
    #　ヒートマップにadd
    a += add_nameshi_mask(opening,img_length)
    

print(a)
thresh = 48
plt.imshow(np.where(a > thresh,a,0))
        
```
